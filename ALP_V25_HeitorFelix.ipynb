##estrutura condicional

```python
nota = float(input("digite sua nota: "))

if (nota >= 7):
 print("aprovado")
elif (nota >= 5):
 print("recuperação")
else:
 print("reprovado")
```

4 - crie um algoritimo que receba uma idade do usuário, e depois verifique em qual faixa etária está à idade:

até 12 - você é uma criança!

de 13 à 18 - você é um adolencente!

de 19 à 60 - você é um adulto!

acima de 60 - você é um idoso!

```python
idade = int(input("Digite sua idade: "))

if idade <= 12:
  print("Você é uma criança!")
elif idade <= 18:
  print("Você é um adolescente!")
elif idade <= 60:
 print("Você é um adulto!")
else:
 print("Você é um idoso!")
```
